# EDA and Other Analysis

```python
df.columns
```

```python
df.drop(['VG','RA','SN','TS','FG'],axis=1,inplace=True)
```

```python
columns_to_convert = ['SLP', 'PP', 'VM', 'PM2.5']
# Convert the specified columns to numeric data type (integer)
for col in columns_to_convert:
    # Clean the column by removing non-numeric characters except for periods
    df[col] = df[col].str.replace(r'[^0-9.]', '', regex=True)
    
    # Convert the values to numeric, setting non-convertible values to NaN
    df[col] = pd.to_numeric(df[col], errors='coerce')

    # Replace NaN or non-convertible values with a default value (e.g., -1)
    default_value = -1
    df[col] = df[col].fillna(default_value).astype('int64')

# Now the specified columns have been converted to int64 data type
print(df)
```

```python
df.info()
```

```python
# Combine 'Day', 'Month', and 'Year' columns into a new 'Date' column
df['Date'] = pd.to_datetime(df[['Year', 'Month', 'Day']])

# Convert the 'Date' column to a string data type
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')

# Drop the original 'Day', 'Month', and 'Year' columns
df.drop(['Day', 'Month', 'Year'], axis=1, inplace=True)

# Now the DataFrame contains a single 'Date' column as a string
print(df)
```
